```python
#seuillage_par_fusion
import cv2
import numpy as np
import matplotlib.pyplot as plt

def seuillage_par_fusion(image, seuil_bas, seuil_haut):
    # Appliquer deux seuillages binaires pour obtenir deux masques
    _, masque_bas = cv2.threshold(image, seuil_bas, 255, cv2.THRESH_BINARY)
    _, masque_haut = cv2.threshold(image, seuil_haut, 255, cv2.THRESH_BINARY_INV)
    
    # Fusionner les deux masques pour obtenir la rgion d'intrt
    masque_fusion = cv2.bitwise_and(masque_bas, masque_haut)
    
    # Optionnel: Appliquer une opration morphologique pour nettoyer le masque
    kernel = np.ones((5,5),np.uint8)
    masque_fusion = cv2.morphologyEx(masque_fusion, cv2.MORPH_OPEN, kernel)
    
    return masque_fusion

# Charger l'image en niveaux de gris
image_path = r"C:\Users\Omar\Desktop\skin_diagnosis\melanoma\Melanoma.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if image is None:
    print("Erreur: Impossible de charger l'image. Vrifiez le chemin.")
else:
    # Appliquer le seuillage par fusion
    seuil_bas = 50
    seuil_haut = 150
    masque = seuillage_par_fusion(image, seuil_bas, seuil_haut)

    # Affichage des rsultats
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Image Originale')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(masque, cmap='gray')
    plt.title('Segmentation par Fusion')
    plt.axis('off')

    plt.show()
```

```python
#training excel
import cv2
import numpy as np
import os
from skimage import feature, color
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
def extraire_caracteristiques(image_path):
    image = cv2.imread(image_path)
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Seuillage d'Otsu
    _, image_seuillee = cv2.threshold(image_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    
    # Extraction de caractristiques
    contours, _ = cv2.findContours(image_seuillee, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contour = max(contours, key=cv2.contourArea) if contours else None
    
    if contour is None:
        return None
    
    # Caractristiques de forme
    aire = cv2.contourArea(contour)
    perimetre = cv2.arcLength(contour, True)
    circularite = ((perimetre ** 2) / (4 * np.pi * aire)) if aire > 0 else 0
    
    # Caractristiques de couleur (moyenne HSV)
    image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    masque = np.zeros(image_gray.shape[:2], np.uint8)
    cv2.drawContours(masque, [contour], -1, 255, -1)
    mean_val = cv2.mean(image_hsv, mask=masque)[:3]
    
    return {
        'Aire': aire,
        'Perimeter': perimetre,
        'Circularity': circularite,
        'Moyenne H': mean_val[0],
        'Moyenne S': mean_val[1],
        'Moyenne V': mean_val[2]
    }

# Parcourir le dossier et traiter chaque image
dossier_images =Path("C://Users//Omar//Desktop//ISIC-2017_Training_Data")
fichiers_images = [f for f in os.listdir(dossier_images) if f.endswith('.jpg')]

# Crer un DataFrame pour stocker les caractristiques
df_caracteristiques = pd.DataFrame()

for fichier in fichiers_images:
    chemin_complet = os.path.join(dossier_images, fichier)
    caracteristiques = extraire_caracteristiques(chemin_complet)
    if caracteristiques:
        df_caracteristiques = df_caracteristiques.append({**caracteristiques, 'Nom Image': fichier}, ignore_index=True)

# Enregistrer les rsultats dans un fichier Excel
nom_fichier_excel = 'Caracteristiques_Images_omar.xlsx'
df_caracteristiques.to_excel(nom_fichier_excel, index=False)

print("Les caractristiques ont t extraites et enregistres dans", nom_fichier_excel)
```
